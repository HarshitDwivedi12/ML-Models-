# This data is Highly Unballanced 
## where,
## 0 represents normal transaction transactions 
## 1 represents fraudulant transaction 

```python
# seprating two data sets for analysis 
legit = credit_card_df[credit_card_df.Class== 0]
fraud = credit_card_df[credit_card_df.Class== 1]
```

```python
print (legit.shape)
print (fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

##  As we got to know the data set is unballanced data set we are going to take equal sample 492 random normal transaction and 492 fradulant transaction in our sample which will be good sample data set for this data set

```python
#Normal transactions
legit_sample= legit.sample(n=492)
```

```python
# concantinate two data sets 
```

```python
new_data= pd.concat([fraud,legit_sample],axis =0)
```

```python
# New Concantinate data set where we have equal 
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# spliting the feautures and  targets 
```

```python
x=new_data.drop(columns='Class',axis=1)
y=new_data['Class']
```

```python
print (x.head())
print (y.head())
```

```python
# spliting the data into train data and test data 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y, random_state=2)
```

```python
print (x.shape, x_train.shape ,x_test.shape)
```

```python
# Modelling 
```

```python
# Logistic regresion 
```

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with train data 
```

```python
model.fit(x_train, y_train)
```

```python
# checking accuracy on train data 
x_train_predictions = model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_predictions,y_train)
```

```python
print ("traing data accuracy: ",training_data_accuracy)
```

```python
# checking accuracy on test data 
x_test_predictions = model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_predictions,y_test)
```

```python
print ("test data accuracy: ",testing_data_accuracy)
```
