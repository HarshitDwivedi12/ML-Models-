```python
# impoting all the dependencies Liberies.
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_squared_error
import statsmodels.api as sm
```

# Importing the Data set

```python
house_df=pd.read_csv("D:\\Data Sets\\kc_house_data.csv")
```

* After analyzing the Data Set we found no null values.

* Droping ID, Date, Lat, Long column as they are non affecting variables to the price 

```python
house_df = house_df.drop(["id", "date","lat","long"], axis=1)
```

```python
house_df.head()
```

# Splitting the data set into train and test data 

```python
house_df.columns
```

```python
# Provided feature variables and target variable
feature_columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
                   'waterfront', 'view', 'condition', 'grade', 'sqft_above',
                   'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'sqft_living15',
                   'sqft_lot15']
target_column = 'price'

df = pd.DataFrame(house_df, columns=feature_columns + [target_column])

X = df[feature_columns]
y = df[target_column]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=None)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

# Normalization the data set 

```python
std=StandardScaler()
X_train_std = std.fit_transform(X_train)
X_test_std = std.transform(X_test)
```

# Modelling the data set

```python
model = LinearRegression()
```

```python
model.fit(X_train_std,y_train)
```

```python
y_predict = model.predict(X_test_std)
```

```python
y_predict
```

```python
y_test
```

# Calculation RMSE, MSE and R-squared score

```python
r2 = r2_score(y_test, y_predict)
mse = mean_squared_error(y_test, y_predict)
rmse = mean_squared_error(y_test, y_predict, squared=False)
```

```python
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared score:", r2)
```

## Model Performance Insights:

* Mean Squared Error (MSE):49226924632.362

This shows how far, on average, our predictions are from the actual values.
A higher MSE means our predictions have bigger errors.
In this case, the model's predictions have quite a bit of variation from the actual values.

* Root Mean Squared Error (RMSE): 221871.4146355091

This is like the "average" size of our prediction errors.
A higher RMSE means our predictions are, on average, off by a larger amount.
The given RMSE of 214352.73 means our predictions are off by around 214,352.73 units on average.

* R-squared score: 0.6464141875096092

This tells us how well our model fits the data compared to a simple average.
A higher R-squared score means our model explains more of the variability in the data.
An R-squared of 0.67 means our model explains about 67% of the variability, which is pretty good but not perfect.
